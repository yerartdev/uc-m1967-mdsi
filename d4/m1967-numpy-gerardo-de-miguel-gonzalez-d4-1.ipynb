```
#::GMG::Generamos 20 elementos aleatorios de [0,1)
a = np.random.rand(20)
np.set_printoptions(precision=2)
print('Elementos:', a)
#::GMG::Elegimos un elemento al azar
n = np.random.choice(a,1)
print('Elegido:', n)
#::Definimos la distancia entre x e y
x = lambda x, y : abs(x - y)
#::GMG::el concepto de cercano se asimila a la distancia
cercano = np.vectorize(x)
#::GMG::Eliminamos el elemento elegido de los elementos
b = np.delete(a,np.argmin(cercano(a,n)))
#::GMG::El elemento más cercano será el que tenga la mínima distancia
mas_cercano = np.argmin(cercano(b,n))
print('El elemento más cercano es:', '{0:0.2f}'.format(b[mas_cercano]))
```
